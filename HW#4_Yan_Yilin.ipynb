Question 4: Plot a figure with multiple lines, as shown in figure 2, by simply adding more plt.plot() functions: https://github.com/fangjin/STAT1129/blob/main/Plot3.ipynb
(Hint: create three y array, add three plot.plot() functions, and add X labels and Y labels.)


```python
A=np.array([1,9])
a=np.array([1,9])
B=np.array([2,3,5,8])
b=np.array([2,6,10,4])
C=np.array([2,5,9])
c=np.array([2,2,2])

plt.plot(A,a,'r')
plt.plot(B,b,'b')
plt.plot(C,c,'g')
plt.xlabel('X Lable')
plt.ylabel('Y Lable')
plt.show()
```

```python
# Or we can use another way to do this: we can chose not to specify the values for x direction, just specify y
# We can use the example of y_point or any defined array such as A/a/B/b to demonstrate this

y_point=np.array([2,8,3,10])

plt.plot(A,'r')
plt.plot(B,'b')
plt.plot(C,'g')
plt.plot(y_point,'y')
plt.xlabel('X Lable')
plt.ylabel('Y Lable')
plt.show()
```

Practice subplots() function, design your own subplot. Using whatever we learned so far, you can draw multiple plots in one figure. Do not copy our class practice code.

```python
#Plot 1: 
A=np.array([1,9])
a=np.array([1,9])
plt.subplot(2,2,1)
plt.plot(A,a,'r')
#this Figure has 2 row, 2 columns, and this is the 1st plot

#Plot 2:
B=np.array([2,3,5,8])
b=np.array([2,6,10,4])
plt.subplot(2,2,2)
plt.plot(B,b,'b')
#this Figure has 2 row, 2 columns, and this is the 2nd plot

#Plot 3:
C=np.array([2,5,9])
c=np.array([2,2,2])
plt.subplot(2,2,3)
plt.plot(C,c,'g')
#this Figure has 2 row, 2 columns, and this is the 3rd plot

plt.show()
```
