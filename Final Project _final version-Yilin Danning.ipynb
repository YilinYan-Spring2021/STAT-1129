```python
# Rock Paper Scissors
# Win the whole Game when wining the two rounds first
import random
import math

class Ai_make_choice: # a random function to generate rock, paper, or scissors for Ai 
    def __init__(self):
        choice_list = ['rock','paper','scissors']
        self.choice = random.choice(choice_list)

def check_validity(choice):      # the valid function: check the validity of the input
        if (choice == "rock" or choice == "paper" or choice == "scissors"):
            return True
        else:  # when the input is not following the requirement(i.e. not valid), ask to retype 
            if (choice != "  "): 
                print("The input is not valid, it should be one of 'rock', 'paper', 'scissors', please retype")
            return False

def check_winner(ai,player):    # the result function,take in ai and player's choice, check winner for each round
    
    if (ai == player):                # they have the same choice
        print("You and Ai have same choice:", ai ,", nobody win or lose.")
        #return "Draw"
    elif ((ai == "rock" and player == "paper")or(ai == "paper" and player == "rock")):
        if (player == "paper"):
            print("Paper covers Rock!! You win this round!")
            return True               # true means the player wins
        else:
            print("Rock cannot covers Paper!! You lose this round...")
            return False              # false means ai wins
    elif ((ai == "papper" and player == "scissors")or(ai == "scissors" and player == "paper")):
        if (player == "scissors"):
            print("Scissors cuts Paper!! You win this round!")
            return True
        else:
            print("Paper cannot cuts Scissors!! You lose this round...")
            return False
    else: #((ai == "rock" and player == "scissors")or(ai == "scissors" and player == "rock")):
        if (player == "rock"):
            print("Rock crushes Scissors!! You win this round!")
            return True
        else:
            print("Scissors cannot crushes Rock!! You lose this round...")
            return False

def scorekeeper(ai_win,player_win,result):   # the scorekepper function: to keep track of the score 
    if (result == True):
        player_win +=1
        return ai_win,player_win
    if (result == False):
        ai_win +=1
        return ai_win,player_win
    else:
        return ai_win,player_win
    
    
def main():   #The main function call the above functions
    print("/////////////// Rock! Paper! Scissors! ///////////////")
    replay = "yes"
    while (replay == "yes"):
        ai_win = 0
        player_win = 0              
        choice = "  "
        r = 1
        
        while(ai_win != 2 and player_win != 2):
            while(not check_validity(choice)):
                print("--------------- Round:",r,"---------------")
                choice = input("Please enter your choice: (rock/paper/scissors)")
            #END while(not check_validity(choice))
            
            ai = Ai_make_choice()
            print("Ai's choice:",ai.choice)
            
            result = check_winner(ai.choice,choice) #Return True or False, then pass this result into scorekeeper
            ai_win,player_win = scorekeeper(ai_win,player_win,result)
            print("Score: ")
            print("Ai: [",ai_win,"/2 ]")
            print("You: [",player_win,"/2 ]")
            print()
            r+=1
            choice = "  "
        # END while(replay == "yes")
         
        if (ai_win ==2): # Determine the final winner
            print("You LOSE the game since ai has already won 2 rounds.")
        else:
            print("You've won 2 rounds, You WIN the game, Congratulation!!!")
            
        replay = input("Play again? (yes/no)")
        while (replay != "yes" and replay != "no"): # when the input is not following the requirement, ask to retype
            print("The input is not valid, it should be one of 'yes' or 'no', please retype.")
            replay = input("Play again? (yes/no)")
        if (replay == "yes"):
            print('                   NEW ROUND')
            print('/////////////// Rock! Paper! Scissors! ///////////////')
    #END: while(replay == "yes")
    print()
    print("Thank you for playing :)")
    

               
if __name__ == "__main__":  # Call to main
    main()
```
