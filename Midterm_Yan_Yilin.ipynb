# Question 1 (6 points): 
Counting the frequencies in a list using dictionary in Python
Given an unsorted list of some elements
[2,4,6,8,4,5,2,1,9,0,4,6,7,4,3,2,1,9,10,3,7,9,6,0,1,3,5,6,7,8,9,10,2,3,6,8,9,10,6,7,4,3], 

- Find the frequency of each distinct element in the list using a dictionary. 
- Plot the frequency results, e.g. using bar chart. 
- Store this dictionary results into json file. 

```python
# First, find the frequency of each distinct element in the list using a dictionary
def FindtheFrequency(MyList):
    
    ## create an empty dictionary
    freq={}
    for items in MyList:
        freq[items]=MyList.count(items)
    
    for key,value in freq.items():
        print("%d:%d"%(key,value))

if __name__=="__main__":
    MyList=[2,4,6,8,4,5,2,1,9,0,4,6,7,4,3,2,1,9,10,3,7,9,6,0,1,3,5,6,7,8,9,10,2,3,6,8,9,10,6,7,4,3]
    FindtheFrequency(MyList)
```

```python
# Then, plot the frequency results, e.g. using bar chart
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


My_List={2:4,4:5,6:6,8:3,5:2,1:3,9:5,0:2,7:4,3:5,10:3}

a=My_List.keys()
b=My_List.values()

df=pd.DataFrame({"x-axis":a,"y-axis":b})

sns.barplot("x-axis","y-axis",palette="RdBu_r",data=df)

plt.show()
```

```python
# Store the dictionary results into a JSON file
import json

## a Python object (dict):
My_List={2:4,4:5,6:6,8:3,5:2,1:3,9:5,0:2,7:4,3:5,10:3}

## dump the python dictionary into a Json file
with open("Midterm_Json_output.json","w") as out:
    json.dump(My_List,out)
```

```python
## After we store this into a json file, we can read it:
f = open("Midterm_Json_output.json","r")
f.read()
```

```python
### Only using this above cell can give us the average, and the biggest spend,
### but I want some explanations and make it clearer, so I add the following.

## Average Spend on Amazon:
MyAverage=df["Total Charged"].mean()
print("My average spend on Amazon is",MyAverage,"dollars.")

## Biggest Purchase
MyBig=df["Total Charged"].max()
print("My biggest purchase on amazon cost me",MyBig,"dollars.")

## Smallest Purchase
MySmall=df["Total Charged"].min()
print("My smallest purchase on amazon only cost me",MySmall,"dollars.")
```
